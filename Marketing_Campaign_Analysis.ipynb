```python
import pandas as pd

df = pd.read_csv('/content/marketing_campaign.csv', sep='\t')

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['Year_Birth'], bins=20)
plt.title('Customer Birth Year Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Income'], bins=30)
plt.title('Income Distribution')
plt.show()
```

```python
campaign_cols = ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3',
                 'AcceptedCmp4','AcceptedCmp5']

df[campaign_cols].sum().plot(kind='bar')
plt.title('Campaign Acceptance')
plt.show()
```

```python
products = ['MntWines','MntFruits','MntMeatProducts',
            'MntFishProducts','MntSweetProducts',
            'MntGoldProds']

df[products].sum().plot(kind='bar')
plt.title('Product Category Sales')
plt.show()
```

```python
funnel = {
    'Customers': len(df),
    'Campaign Responses': df['Response'].sum()
}

plt.figure(figsize=(8,5))
bars = plt.bar(funnel.keys(), funnel.values())

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y + 20, int(y),
             ha='center')

plt.title("Marketing Funnel")
plt.ylabel("Number of Customers")
plt.show()
```

Buisness Insights
1. Wine products generate the highest revenue among all product categories.

2. Campaign 5 achieved the best customer response rate.

3. High-income customers demonstrate greater purchasing power and engagement.

4. The overall campaign conversion rate is relatively low, suggesting opportunities for better targeting.

5. Premium customer segments should receive personalized offers to improve ROI.
